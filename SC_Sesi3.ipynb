```python
from scipy.interpolate import interp1d  #function interpolasi
from scipy.interpolate import griddata #function cubic spline
import numpy as np #function operasi numerik
import matplotlib.pyplot as plt #untuk membuat grafik
plt.style.use ('ggplot') #style gambar grafik
```

Linear Interpolasi

```python
x = [0, 1, 2]
y = [1, 3, 2]
```

```python
f = interp1d(x, y)
y_ask = f(1.5)
print(y_ask)
```

```python
plt.figure (figsize = (10, 8)) #ini untuk size yang kita mau
plt.plot (x, y, '-ob') # - itu lurus, o untuk bentuk titik, klo b itu warna dengan variabel pertamanya
plt.plot (1.5, y_ask, 'ro') #untuk nilai ini untuk lingkaran doang
plt.title ('Linear Interpolation at x = 1.5')
plt.xlabel ('x')
plt.ylabel ('y')
plt.show
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 3, 5, 2, 2]
```

```python
f = interp1d (x,y)

x_lin = np. linspace (0, 5, 100) #membuat 100 nilai yang tersebar merata antara 0 hingga 5 berjumlah 100
y_lin = f(x_lin) # Menghitung nilai y untuk semua x baru

y_ask = f(3.1) #mau cari titik dari 3.1 nanti tinggal diprint aja
print (y_ask)
```

Newton Interpolation

```python
%matplotlib inline
def divided_diff (x,y):

  n = len (y)
  coef = np.zeros ([n,n])
  #the first column is y
  coef [:,0] = y

  for j in range (1,n):
    for i in range (n-j):
      coef [i][j] = \
      (coef[i+1][j-1]-coef [i][j-1]) / (x[i+j]-x[i])

  return coef

def newton_poly(coef, x_data, x): # Define newton_poly outside divided_diff
  n = len (x_data) - 1
  p = coef [n]
  for k in range (1, n+1):
    p = coef [n-k] + (x - x_data [n-k]) *p
  return p # Return outside the loop
```

```python
x = np.array ([-5, -1, 0, 2])
y = np.array ([-2, 6, 1, 3])

a_s = divided_diff(x,y)[0,:]

x_new = np.arange (-5, 2.1, .1)
y_new = newton_poly (a_s, x, x_new)

plt.figure (figsize = (12, 8))
plt.plot (x, y, 'bo')
plt.plot (x_new, x_new, 'go')
plt.plot (x_new, y_new)
```
