```python
import numpy as np
import matplotlib.pyplot as plt
# Corrected the style name by adding a hyphen
plt.style.use('seaborn-v0_8-poster')

f = lambda t,s : np.exp(-t)
h = 0.1
t_min = 0
t_max = 1

t = np.arange(t_min,t_max+h,h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0,len(t)-1):
    s[i+1] = s[i] + h*f(t[i],s[i])

plt.figure(figsize = (12,8))
plt.plot(t,s, 'bo--', label = 'approx(euler h=0.1)')
plt.plot(t, -np.exp(-t), 'g', label = 'exact solution')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi turunan dari ODE y' = x^2 - 4y
def f(x, y):
    return x**2 - 4*y

# Parameter
h = 0.01  # step size
t = np.arange(0, 1 + h, h)  # grid
y0 = 1  # initial condition

# Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(len(t) - 1):
    y[i+1] = y[i] + h * f(t[i], y[i])

# Solusi Eksak (analitik)
# Hasil dari menyelesaikan y' + 4y = x^2
# y(x) = (1/4)x^2 - (1/8)x + (1/32) + (31/32)*e^(-4x)
exact_y = (1/16)* (1-np.exp (-4*t)) * (t*3) / 12 - (t*2) / 8 + t / 32

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate (Euler)')
plt.plot(t, exact_y, 'g', label='Exact')
plt.title("Approximate and Exact Solution \n for the ODE $y' + 4y = x^2$")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid
plt.legend(loc='lower right')
plt.show()
```

```python
#2. y' = sin y, y(0) = 1

import numpy as np
import matplotlib.pyplot as plt

# Fungsi dari turunan
def f(x, y):
    return np.sin(y)

# Parameter
h = 0.01  # step size
x = np.arange(0, 5 + h, h)
y = np.zeros(len(x))
y[0] = 1

# Metode Euler
for i in range(len(x) - 1):
    y[i+1] = y[i] + h * f(x[i], y[i])

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'b', label="Approximate (Euler's Method)")
plt.title("Approximate Solution for $y' = \sin(y),\ y(0) = 1$")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.legend()
plt.show()


```
