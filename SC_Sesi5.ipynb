```python
import numpy as np

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    T = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    return T

# Fungsi sin(x)
f = np.sin
a = 0
b = np.pi

# Hasil untuk n=8
result_8 = trapezoidal_rule(f, a, b, 8)
# Hasil untuk n=16
result_16 = trapezoidal_rule(f, a, b, 16)

# Cetak hasil
print(f"Hasil integral dengan 8 panel  : {result_8:.6f}")
print(f"Hasil integral dengan 16 panel : {result_16:.6f}")
```
